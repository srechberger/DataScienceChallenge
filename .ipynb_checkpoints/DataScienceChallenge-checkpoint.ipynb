# <span style="color:#0B0B61">1. Business and Data Understanding</span>

Grundsätzlich stehen zur Befriedigung des Energiebedarfs im Unternehmen 3 Quellen zur Verfügung: 
* Photovoltaik-Anlage (PV)
* Batteriesystem (Battery)
* Netz (Netz)

Auf der Gegenseite stehen sämtliche Geräte (Devices), die Energie beziehen.

**Primäre Zielsetzung:** Optimierung des Energieverbrauches der PV-Anlage erzeugten Energiemenge

Ziel ist es möglichst wenig Energie aus dem Netz beziehen zu müssen. Hierbei ist es jedoch auch wichtig etwaige differierenden Stromtarife zu berücksichtigen, sprich potenziell herrschen je nach Tageszeit unterschiedliche Preise vor. Darüber hinaus ist zu berücksichtigen, dass möglicherweise bei Überschreitung von vereinbarten Energiekonsumspitzen die Kosten pro kWh signifikant ansteigen. Folglich gilt es die selbsterzeugte Energie (PV-Strom) zielgerichtet einzusetzen.

Die Verteilung des Strombezugs wird über den Controller (Steuerelement) optimiert. Die Energie aus der PV-Anlage kann entweder direkt an die Verbraucher fließen oder zur Speicherung (Battery) verwendet werden. Falls der Speicher voll ist und die Verbraucher zur Gänze durch die PV-Anlage versorgt werden können, so wird der Energieüberschuss in das Netz eingespeißt. Bei dem aus der PV-Anlage erzeugten Strom handelt es sich um Gleichstrom, der mittels Wechselrichter (Inverter) in Wechselstrom umgewandelt wird.

Im Allgemeinen können die Verbraucher die benötigte Energie aus dem Netz (Grid), PV-Anlage (PV) und/oder dem Speicher (Battery) beziehen. In der unten angeführten Grafik werden die gegebenen Variablen für das Leistungsvermögen (POWER_\* bzw. Watt) und die fließende Engergiemenge (ENERGY_\* bzw. Wattstunden) illustriert.

<img src="images/Business_Understanding.png">

Wie bereits kurz erläutert ist bei den Variablen zwischen Leistung (Leistungsvermögen, Power) und Energie (Energiemenge, Energy) zu differenzieren. Die Leistung wird in Form von Watt gemessen. Watt (W) ist eine Einheit für die Leistung, und Leistung ist die Geschwindigkeit, mit der Energie erzeugt oder verbraucht wird. Eine Wattstunde (Wh) ist eine Einheit der Energie, eine Methode zur Messung der Menge der geleisteten bzw. erzeugten Arbeit. Mit Wattstunden (Wh) wird die Menge an Energie für einen bestimmten Zeitraum gemessen und mit Watt die Leistungsabgabe (Leistungsgeschwindigkeit) zu einem bestimmten Zeitpunkt dargestellt.

Das vorliegende Datenset beinhaltet insgesamt 11 Attribute die zur Verfügung stehen. Die Variable "Time" stellt hierbei den Messungszeitpunkt dar. Die anderen 10 Variablen stellen die aufgezeichneten Messwerte für Leistung und Energiemenge je Komponente dar. In der folgenden Tabelle werden Annahmen getroffen welche Messwerte im Unternehmen aufgezeichnet wurden:

|Attribut|Beschreibung|
|--------|--------|
|**ENERGY_CHARGED**|Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) in den Speicher (Battery) eingespeißt wurde|
|**ENERGY_DISCHARGED**|Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) vom Speicher (Battery) ausgeliefert wurde|
|**ENERGY_IMPORTED**|Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) vom Netz (Grid) bezogen wurde|
|**ENERGY_EXPORTED**|Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) in das Netz (Grid) eingespeißt wurde|
|**ENERGY_PRODUCED**|Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) von der PV-Anlage produziert wurde|
|**ENERGY_CONSUMED_CALC**|Gesamte Energiemenge in Wattstunden (Wh), welche in einem Zeitraum (15 min) von den Geräten (Devices) konsumiert wurde|
|**POWER_STORAGE**|Leistung in Watt (W), welche zu einem bestimmten Zeitpunkt im Speicher vorhanden war|
|**POWER_GRID**|Leistung in Watt (W), welche zu einem bestimmten Zeitpunkt vom Netz zur Verfügung stand|
|**POWER_PRODUCTION**|Leistung in Watt (W), welche zu einem bestimmten Zeitpunkt von der PV-Anlage produziert wurde|
|**POWER_CONSUMPTION**|Leistung in Watt (W), welche zu einem bestimmten Zeitpunkt von den Geräten abgerufen wurde|

## <span style="color:#0404B4">2.1. Data Cleaning</span>

Die Rohdaten werden mittels pandas-Bibliothek in ein Dataframe geladen:

```python
import pandas as pd

# Fetch data
data_file_path = "data/ntuity_datascience_challenge.csv"
data = pd.read_csv(data_file_path)
```

### 2.1.1. Überprüfung der Datentypen und Durchführung notwendiger Type-Casts

```python
# Inspect datatypes of all cols
print(data.dtypes)

# Transform time column from object to datetime
data['Time'] = pd.to_datetime(data['Time'])
```

### 2.1.2. Überprüfung auf fehlende Werte (null)

```python
# Set display options
pd.set_option('display.max_columns', 20)

# Check missing values per column
data.isnull().sum()

data.head
```

### 2.1.3. Feature Engineering

Es wird die Annahme getroffen, dass die Stromproduktion durch die PV-Analage vom Bewölkungsgrad und der Tageszeit abhängig ist. Für die Abhängigkeit zwischen PV-Produktion und Bewölkungsgrad sowie Tageslicht müssten externe Datenquellen angebunden werden, sprich eine notwendige Datenanreicherung müsste erfolgen. Hinsichtlich der angenommenen Abhängigkeit wird vorsorglich eine separate Variable **DayTime** angelegt, die die Uhrzeitbeinhaltet.

```python
# Add new column with DayTime from Time column
hour = data['Time'].dt.hour
minute = data['Time'].dt.minute
data['DayTime'] = hour * 100 + minute
```

### 2.1.4. Validierung der Wertebereiche sämtlicher Attribute (inkl. Outliers):

Zur Überprüfung der Werte wird eine Leistungsbilanz-Variable **POWER_BALANCE** erstellt, um validieren zu können ob das bereitgestellte Leistungsvermögen zum Zeitpunkt der Aufzeichnung mit den verbrauchten Leistungen in einem ausgeglichenen Verhältnis steht.

```python
# Create new column 'POWER_BALANCE'
data['POWER_BALANCE'] = data['POWER_PRODUCTION'] + data['POWER_STORAGE'] + data['POWER_GRID'] - data['POWER_CONSUMPTION']
```

Zur Überprüfung der Wertebereiche werden im unten angeführten Abschnitt sämtliche Quartile sowie übliche Kennzahlen (Anzahl, Mittelwert, Standardabweichung, Min, Max).

```python
# Generate overview for all attributes
data.describe().apply(lambda s: s.apply(lambda x: format(x, 'g')))
```

```python
# Check exponential values in detail
print(data['ENERGY_PRODUCED'].describe().apply(lambda x: format(x, 'f')))
print("-------------------")
print(data['ENERGY_CONSUMED_CALC'].describe().apply(lambda x: format(x, 'f')))
print("-------------------")
print(data['ENERGY_EXPORTED'].describe().apply(lambda x: format(x, 'f')))
print("-------------------")
print(data['ENERGY_IMPORTED'].describe().apply(lambda x: format(x, 'f')))
print("-------------------")
print(data['POWER_BALANCE'].describe().apply(lambda x: format(x, 'f')))
```

**Analyse der Wertebereiche**

Die Ergebnisse zeigen, dass das Datenset bei den Variablen **ENERGY_CHARGED** und **ENERGY_DISCHARGED** ausschließlich 0-Werte beinhaltet. Die Variable **POWER_STORAGE** weist durchwegs Veränderungen über den Beobachtungszeitraum auf. Resultierend sollten sowohl eingehende als auch ausgehende Stromflüsse von der Batterie geflossen sein. Möglicherweise sind beim Speicher (Battery) die Zähler defekt und es werden folglich keine Aufzeichnungen geliefert. 

Die Variablen **POWER_STORAGE** und **POWER_GRID** weisen teilweise negative Werte auf. Diese werden als ausgehende Energieflüsse interpretiert.

# <span style="color:#0B0B61">3. Modeling</span>

Ziel ist es ein einfaches Vorhersagemodell zu erstellen, das die Produktion der PV-Anlage für die nächsten beiden Tage prognostiziert. Zur Erstellung der Vorhersagemodelle und entsprechender Evaluierung werden die Daten in Training und Testdaten unterteilt. Hierzu wird eine 50-50-Split (50 % Trainingsdaten und 50 % Testdaten) verwendet. Der Beobachtungszeitraum erstreckt sich vom 12. Oktober 2020 bis 25. Oktober 2020. Folglich werden die Zeitintervalle wie folgt definiert:
* Trainingsintervall: 12.10.2020-19.10.2020
* Testintervall: 20.10.2020-25.10.2020

```python
import datetime

# Set Date as index
data = data.set_index('Time')

# Define intervals
date_train_until = datetime.datetime(2020, 10, 19, 23, 59)

# Split train and test
train = data.loc[(data.index <= date_train_until)]
test = data.loc[(data.index > date_train_until)]

# Create target object
y_train = train.POWER_PRODUCTION
y_test = test.POWER_PRODUCTION

# Select features
features = ['DayTime']
X_train = train[features]
X_test = test[features]

```

Für die Modelle könnten lediglich die DayTime-Variable und die Time-Variable (index) relevant sein. Die restlichen Variablen stehen zwar in gewisser Art und Weise in Abhängikeit zur PV-Produktion, jedoch beeinflusst die PV-Anlage die restlichen Komponenten nur unidirektional und ist nicht von Netz, Batterie oder Verbrauch abhängig.

Darüber hinaus könnten externe Datenquellen in Betracht gezogen werden, um bessere Prognoseergebnisse erzielen zu können. Beispielsweise würde es sich anbieten die Sonnenaufgänge und -untergänge je Datum zu berücksichtigen, um mittels Boolean die Korrelation zur Sonneneinstrahlung abbilden zu können. Zusätzlich können auch Wetterforecasts herangezogen werden, um den Bewölkungsgrad berücksichtigen zu können.

Für diese Aufgabe wird lediglich ein RandomForestRegressor als Methode (Ensemble Methods - Bagging) verwendet, um eine exemplarische Modellentwicklung darstellen zu können. Hierbei wird ein Grid-Search zur Hyperparameter-Optimierung angewendet.

```python
from models import hyperparameter_rf_gs

# Create RandomForestRegressor model incl. hyperparameter tuning
rf_model = hyperparameter_rf_gs(X_train, X_test, y_train, y_test)
```

# <span style="color:#0B0B61">4. Evaluation</span>

## <span style="color:#0404B4">4.1. Evaluierung anhand der Trainings- und Testdaten</span>

Das erstelle RandomForest-Modell wird nun mittels einer gängigen Fehlerfunktion namens RMSE evaluiert. Hierzu werden einerseits die Trainingsdaten herangezogen, um zu überprüfen inwiefern das Muster anhand der bekannten Daten erkannt wurde. Andererseits wird anhand der Testdaten evaluiert inwiefern das Modell den unbekannten Verlauf (unbekannt im Sinne für das Modell) prognostizieren kann.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Define Function for RMSE
def rmse(x, y):
    return sqrt(mean_squared_error(x, y))

# Prediction training data (pattern recognition)
y_train_predicted = rf_model.predict(X_train)

# Evaluation training data (pattern recognition)
print("Training Evaluation")
print("RMSE:", rmse(y_train, y_train_predicted))

# Prediction test data
y_test_predicted = rf_model.predict(X_test)

# Evaluation test data
print("Test Evaluation")
print("RMSE:", rmse(y_test, y_test_predicted))
```

```python
from plotcharts import plot_evaluation

train_predictions = pd.DataFrame([[x, y] for x, y in zip(y_train.index, y_train_predicted)], columns=["Time", "Predictions"])
train_predictions = train_predictions.set_index('Time')

test_predictions = pd.DataFrame([[x, y] for x, y in zip(y_test.index, y_test_predicted)], columns=["Time", "Predictions"])
test_predictions = test_predictions.set_index('Time')

plot_evaluation(y_train, train_predictions, y_test, test_predictions)
```

## <span style="color:#0404B4">4.2. Prognose für 2 Tage</span>

Das Datenset erstreckt sich über einen Zeitraum von 12.10.2020 bis 25.10.2020. Es wird nun eine Vorhersage für die nächsten beiden Tage erstellt (26.10 + 27.10.).

```python
from plotcharts import plot_predictions
from datetime import timedelta

# Get actual values
powerProduction = data.POWER_PRODUCTION

# Prepare dataframe to get prediction horizont
start_time = datetime.datetime(2020, 10, 26, 0, 0)
end_time = datetime.datetime(2020, 10, 27, 23, 59)
curr_time = start_time

rows = []
i = 0

while curr_time <= end_time:
    rows.append([i, curr_time])
    i += 1
    curr_time = curr_time + timedelta(minutes=15)

prediction_horizont = pd.DataFrame(rows, columns=["Row", "Time"])
prediction_horizont = prediction_horizont.drop('Row', axis=1)

hour = prediction_horizont['Time'].dt.hour
minute = prediction_horizont['Time'].dt.minute
prediction_horizont['DayTime'] = hour * 100 + minute
prediction_horizont = prediction_horizont.set_index('Time')

# Predict
predictions = rf_model.predict(prediction_horizont)

predictions = pd.DataFrame([[x, y] for x, y in zip(prediction_horizont.index, predictions)], columns=["Time", "Predictions"])
predictions = predictions.set_index('Time')

# Plot results
plot_predictions(powerProduction, predictions)
```
